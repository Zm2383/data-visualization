```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
#functional
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'-r')

plt.subplot(1,2,2)
plt.plot(y,x,'-b')
```

```python
#oo
fig=plt.figure()

axes=fig.add_axes([0.1,0,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y')
```

```python
fig= plt.figure() #create empty figure

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.5,0.15,0.35,0.3]) #location
axes1.plot(x,y)
axes2.plot(y,x)
axes2.set_title('small')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)


#axes.plot(x,y)
plt.tight_layout()

for current_ax in axes: #axes is a array of matplot
    current_ax.plot(x,y)
```

```python
axes
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('First')
axes[1].plot(y,x)
axes[1].set_title('Second')
```

# Figure Size And DPI




```python
fig=plt.figure(figsize=(8,2),dpi=100)

ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.plot(y,x)
ax.set_title('Title')
ax.set_ylabel('Y label')
ax.set_xlabel('X label')
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')

ax.legend(loc=0) #legend method need to be set in label above


ax.set_title('Title')
ax.set_ylabel('Y label')
ax.set_xlabel('X label')
```

# Custimising Plot

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,color='orange',linewidth=1 ,alpha=0.5,linestyle='-.',marker='o',markersize=10
       ,markerfacecolor='yellow',markeredgewidth=4,markeredgecolor='blue') #RGB Hex Code,linestyle=ls
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,color='orange',linewidth=2 ,alpha=0.5,linestyle='--') #RGB Hex Code,linestyle=ls
ax.set_xlim([0,1])
ax.set_ylim([0,2])#set limtation
```
